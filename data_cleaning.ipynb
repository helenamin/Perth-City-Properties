```python
import pandas as pd
from datetime import datetime
```

```python
perth = pd.read_csv("static/data/Perth.csv")
west_perth = pd.read_csv("static/data/WestPerth.csv")
east_perth = pd.read_csv("static/data/EastPerth.csv")
crawley = pd.read_csv("static/data/Crawley.csv")
nedlands = pd.read_csv("static/data/Nedlands.csv")
northbridge = pd.read_csv("static/data/Northbridge.csv")

northbridge
```

```python
frames = [perth, west_perth, east_perth, crawley, nedlands, northbridge]

cityofperth_df = pd.concat(frames)

cityofperth_df
```

```python
cityofperth_df = cityofperth_df.dropna()

cityofperth_df
```

```python
#to remove '$' from price
cityofperth_df['price'] = cityofperth_df['price'].map(lambda x: x.lstrip('$'))
```

```python
#to remove ',' from price
cityofperth_df['price'] = cityofperth_df['price'].str.replace(',', '')
```

```python
cityofperth_df['price']
```

```python
#to remove '$' and 'pw' from rent
cityofperth_df['rent'] = cityofperth_df['rent'].map(lambda x: x.lstrip('$').rstrip('pw'))
```

```python
cityofperth_df['rent']
```

```python
cityofperth_df.dtypes
```

```python
cityofperth_df['address'] = cityofperth_df['address'].astype('string')
cityofperth_df['suburb'] = cityofperth_df['suburb'].astype('string')
cityofperth_df['state'] = cityofperth_df['state'].astype('string')
cityofperth_df['full_address'] = cityofperth_df['full_address'].astype('string')
cityofperth_df['type'] = cityofperth_df['type'].astype('string')
cityofperth_df['price'] = cityofperth_df['price'].astype('float64')
cityofperth_df['sold date'] = cityofperth_df['sold date'].astype('datetime64[ns]')
cityofperth_df['Land size'] = cityofperth_df['Land size'].astype('float64')
cityofperth_df['Year Built'] = cityofperth_df['Year Built'].astype('datetime64[ns]')
cityofperth_df['rent'] = cityofperth_df['rent'].astype('float64')
cityofperth_df['rent date'] = cityofperth_df['rent date'].astype('datetime64[ns]')
cityofperth_df['agent'] = cityofperth_df['agent'].astype('string')

cityofperth_df.dtypes
```
